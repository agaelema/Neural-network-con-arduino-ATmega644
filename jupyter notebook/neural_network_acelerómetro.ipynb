```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("datos_aceler√≥metro.csv")
```

```python
df[:10]
```

```python
df.shape
```

```python
datos= df.values[:,:15]
etiquetas= df.values[:,15]
```

```python
#los dato deben estar "normalizados" entes valores de 0 a 1
datos=datos/255
```

```python
datos[:5]
```

```python
encoder = LabelEncoder()
encoder.fit(etiquetas)
encoded_Y = encoder.transform(etiquetas)
etiquetas_ = np_utils.to_categorical(encoded_Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(datos, etiquetas_, test_size=0.2, random_state=42)
```

```python
# Datos para la validaci√≥n
print(X_test[:10])
print(y_test[:10])
```

```python
model = Sequential()
model.add(Dense(16, input_dim=15, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(3, activation='sigmoid'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=300, batch_size=5)
```

```python
scores = model.evaluate(X_test, y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# funciones para escribir los pesos y sesgos(bias)

# capa : numero de capa a escribir
# Num_Dim : dimensi√≥n de la capa de entrada
# num_Neuronas : cantidad de neuronas de la capa
# modelo :  matriz con los pesos y sesgos de la capa (esta variable no tiene nombre descriptivo).

def escribir_pesos(capa, Num_Dim, num_Neuronas, modelo):

    salida =""
    print("// Weight of Layer: " + str(capa) )
    salida = "mtx_type WL" + str(capa) + "["+ str(Num_Dim) +"]["+ str(num_Neuronas) + "] = {\n"
    for u in range(0, Num_Dim+1): 
        if(u < 14):
            print(salida)
            salida = "" 
        elif(u == Num_Dim):
            salida += "};"
            print(salida)
            break
        for i in range(0, num_Neuronas): 
            if (i == 0):
                salida += "{ "
                salida += str(modelo[0][u][i])
                salida += ", "
            elif(i == num_Neuronas-1):
                salida += str(modelo[0][u][i])
                salida += " },\n "
            elif(i%5==0):
                salida += str(modelo[0][u][i])
                salida += ",\n"
            else:
                salida += str(modelo[0][u][i])
                salida += ", "
                               
def escribe_bias(capa, num_Neuronas, modelo):

    salida =""
    print("// bias of Layer: " + str(capa) )
    salida = "mtx_type BL" + str(capa)+ "[1]["+ str(num_Neuronas)+ "] = {\n"
    for u in range(0, num_Neuronas): 
        if(u < num_Neuronas-1):
            salida += str(modelo[1][u])
            salida += ", "    
        elif(u == num_Neuronas-1):
            salida += str(modelo[1][u]) 

    salida += "};"        
    print(salida)
```

```python
# capa 0
a =  model.layers[0].get_weights()
escribir_pesos(0, 15, 16, a)
escribe_bias(0, 16, a)
```

```python
# capa 1
b =  model.layers[1].get_weights()
escribir_pesos(1, 16, 16, b)
escribe_bias(1, 16, b)
```

```python
# capa 2
c =  model.layers[2].get_weights()
escribir_pesos(2, 16, 3, c)
escribe_bias(2, 3, c)
```
